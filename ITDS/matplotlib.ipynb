```python
import pandas as pd

# Sample data
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
    'Sales': [5000, 6000, 7000, 8000, 7500, 8500, 9500, 9000, 8000, 8500, 7500, 9000],
    'Customers': [45, 50, 55, 60, 58, 65, 70, 68, 63, 66, 60, 70],
    'Expenses': [2000, 2500, 3000, 3200, 3100, 3300, 3500, 3400, 3100, 3200, 3000, 3400],
    'Product_A_Sold': [15, 20, 25, 30, 28, 35, 38, 37, 32, 34, 29, 36],
    'Product_B_Sold': [30, 35, 40, 45, 42, 48, 50, 46, 44, 46, 42, 47]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Saving the DataFrame as a CSV file
df.to_csv('sales_data.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Data
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sales = [5000, 6000, 7000, 8000, 7500, 8500, 9500, 9000, 8000, 8500, 7500, 9000]
customers = [45, 50, 55, 60, 58, 65, 70, 68, 63, 66, 60, 70]
expenses = [2000, 2500, 3000, 3200, 3100, 3300, 3500, 3400, 3100, 3200, 3000, 3400]
product_a_sold = [15, 20, 25, 30, 28, 35, 38, 37, 32, 34, 29, 36]
product_b_sold = [30, 35, 40, 45, 42, 48, 50, 46, 44, 46, 42, 47]

# 1. Line Plot
plt.figure(figsize=(10, 6))
plt.plot(months, sales, marker='o', linestyle='-', color='b', label='Sales')
plt.xlabel('Months')
plt.ylabel('Sales ($)')
plt.title('Monthly Sales in 2023')
plt.grid(True)
plt.show()

# 2. Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(months, customers, color='orange')
plt.xlabel('Months')
plt.ylabel('Number of Customers')
plt.title('Monthly Customer Count')
plt.show()

# 3. Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(sales, expenses, c=customers, cmap='viridis', marker='x')
plt.xlabel('Sales ($)')
plt.ylabel('Expenses ($)')
plt.title('Sales vs. Expenses')
plt.colorbar(label='Number of Customers')
plt.show()

# 4. Pie Chart
labels = ['Product A', 'Product B']
sizes = [product_a_sold[-1], product_b_sold[-1]]
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Product Sales Distribution in December')
plt.show()

# 5. Subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Line Plot
axs[0, 0].plot(months, sales, marker='o', linestyle='-', color='b')
axs[0, 0].set_title('Monthly Sales')

# Bar Chart
axs[0, 1].bar(months, customers, color='orange')
axs[0, 1].set_title('Customer Count')

# Scatter Plot
sc = axs[1, 0].scatter(sales, expenses, c=customers, cmap='viridis', marker='x')
axs[1, 0].set_title('Sales vs. Expenses')
fig.colorbar(sc, ax=axs[1, 0])

# Pie Chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Sales Distribution (December)')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# 1. Scatter Plots for Feature Relationships
plt.figure(figsize=(10, 6))
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['target'], cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs. Sepal Width')
plt.colorbar(label='Species')
plt.show()

# 2. Pair Plot
sns.pairplot(df, hue='species', markers=["o", "s", "D"])
plt.suptitle('Pair Plot of Iris Features', y=1.02)
plt.show()

# 3. Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.drop(columns=['target']), orient='h')
plt.title('Box Plot of Iris Features')
plt.show()

# 4. Histograms for Feature Distributions
df.drop(columns=['target']).hist(bins=20, figsize=(12, 8), grid=False)
plt.suptitle('Histogram of Iris Features')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(df)

# Visualize the clusters using sepal length and width
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['sepal length (cm)'], y=df['sepal width (cm)'], hue=df['cluster'], palette='viridis')
plt.title('K-means Clustering on Iris Dataset (Sepal)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
df['dbscan_cluster'] = dbscan.fit_predict(df[iris.feature_names])

# Visualize the DBSCAN clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['sepal length (cm)'], y=df['sepal width (cm)'], hue=df['dbscan_cluster'], palette='viridis')
plt.title('DBSCAN Clustering on Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# Apply Agglomerative Clustering
agg_cluster = AgglomerativeClustering(n_clusters=3)
df['agg_cluster'] = agg_cluster.fit_predict(df[iris.feature_names])

# Visualize the Agglomerative clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['sepal length (cm)'], y=df['sepal width (cm)'], hue=df['agg_cluster'], palette='viridis')
plt.title('Agglomerative Clustering on Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans, DBSCAN

# Generate moons dataset
X_moons, y_moons = make_moons(n_samples=300, noise=0.1, random_state=42)

# K-means on moons dataset
kmeans_moons = KMeans(n_clusters=2, random_state=42)
labels_kmeans_moons = kmeans_moons.fit_predict(X_moons)

plt.figure(figsize=(8, 6))
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=labels_kmeans_moons, cmap='viridis')
plt.title('K-means Clustering on Moons Dataset')
plt.show()

# DBSCAN on moons dataset
dbscan_moons = DBSCAN(eps=0.2, min_samples=5)
labels_dbscan_moons = dbscan_moons.fit_predict(X_moons)

plt.figure(figsize=(8, 6))
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=labels_dbscan_moons, cmap='viridis')
plt.title('DBSCAN Clustering on Moons Dataset')
plt.show()
```

```python
from sklearn.datasets import make_circles

# Generate circles dataset
X_circles, y_circles = make_circles(n_samples=300, factor=0.5, noise=0.05)

# K-means on circles dataset
kmeans_circles = KMeans(n_clusters=2, random_state=42)
labels_kmeans_circles = kmeans_circles.fit_predict(X_circles)

plt.figure(figsize=(8, 6))
plt.scatter(X_circles[:, 0], X_circles[:, 1], c=labels_kmeans_circles, cmap='viridis')
plt.title('K-means Clustering on Circles Dataset')
plt.show()

# DBSCAN on circles dataset
dbscan_circles = DBSCAN(eps=0.2, min_samples=5)
labels_dbscan_circles = dbscan_circles.fit_predict(X_circles)

plt.figure(figsize=(8, 6))
plt.scatter(X_circles[:, 0], X_circles[:, 1], c=labels_dbscan_circles, cmap='viridis')
plt.title('DBSCAN Clustering on Circles Dataset')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Function to perform and visualize hierarchical clustering with different linkage types
def hierarchical_clustering(linkage_type):
    # Perform Agglomerative Clustering
    agglom = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage=linkage_type)
    df['cluster'] = agglom.fit_predict(df)

    # Plot the dendrogram
    plt.figure(figsize=(10, 6))
    sch.dendrogram(sch.linkage(df.iloc[:, :-1], method=linkage_type))
    plt.title(f'Dendrogram for {linkage_type.capitalize()} Linkage')
    plt.xlabel('Samples')
    plt.ylabel('Distance')
    plt.show()

    # Visualize the clusters using sepal length and width
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=df['sepal length (cm)'], y=df['sepal width (cm)'], hue=df['cluster'], palette='viridis')
    plt.title(f'Hierarchical Clustering with {linkage_type.capitalize()} Linkage')
    plt.xlabel('Sepal Length (cm)')
    plt.ylabel('Sepal Width (cm)')
    plt.show()

# Perform clustering with different linkage types
linkages = ['single', 'complete', 'average', 'ward']
for linkage in linkages:
    hierarchical_clustering(linkage)

```
